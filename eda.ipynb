```python
# Llegir fitxer saltant les 12 primeres línies i detectant separador
df_weather = pd.read_csv("C:/Users/irene/Desktop/uni 4t carrera/PM/data/data/fabra_diari.txt", sep="\t", skiprows=12)

# Rename les columnes a noms reconeguts per pd.to_datetime
df_weather.rename(columns={"ANY": "year", "MES": "month", "DIA": "day"}, inplace=True)

# Assegurar-nos que year, month, day són enters
df_weather["year"] = df_weather["year"].astype(int)
df_weather["month"] = df_weather["month"].astype(int)
df_weather["day"] = df_weather["day"].astype(int)

# Crear columna Data com a datetime
df_weather["Data"] = pd.to_datetime(df_weather[["year", "month", "day"]])

# Filtrar només anys 2022 a 2024
df_weather = df_weather[(df_weather["year"] >= 2022) & (df_weather["year"] <= 2024)]

# Eliminar columna INS
df_weather = df_weather.drop(columns=["INS"])

df_weather.head()
```

```python
# Dictionary of selected dataset names for our project and their corresponding DataFrames
datasets = {
    "consum": df1,
    "fuites": df3,
    "meteo": df_weather,
    "rendes": df_renda
}

os.makedirs("outputs", exist_ok=True)

# Create a Pandas Excel writer
with pd.ExcelWriter("outputs/data_inventory.xlsx", engine="openpyxl") as writer:
    
    # Iterate over each dataset and write detailed metadata to a separate sheet
    for name, df in datasets.items():
        print(f"Processing {name}...")
        
        # Build metadata dataframe for each dataset
        meta = pd.DataFrame({
            "Column Name": df.columns,
            "Data Type": [str(df[col].dtype) for col in df.columns],
            "Missing Values (%)": [round(df[col].isna().mean() * 100, 2) for col in df.columns],
            "Num Unique Values": [df[col].nunique() for col in df.columns],
            "Example Value": [df[col].dropna().iloc[0] if df[col].notna().any() else None for col in df.columns]
        })

        # Add overall summary in the first rows
        summary = pd.DataFrame({
            "Metric": ["Num Rows", "Num Columns", "Total Missing (%)"],
            "Value": [df.shape[0], df.shape[1], round(df.isna().mean().mean() * 100, 2)]
        })

        # Write both to Excel (summary on top, then metadata)
        summary.to_excel(writer, sheet_name=name, index=False, startrow=0)
        meta.to_excel(writer, sheet_name=name, index=False, startrow=5)
    
    print("Data inventory Excel file created successfully!")
```
